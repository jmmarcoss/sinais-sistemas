\begin{equation}
s[n]=
\begin{cases} 
0 , \text{para } -2 \leq n \leq 4 \\
n , \text{caso contrário} 
\end{cases}
~~~~~~~~~~~~~~~~\text{e}~~~~~~~~~~~~~~~~
s(t) = 
\begin{cases} 
0, \text{para } t < 3 \\[2mm]
1 , \text{caso contrário}
\end{cases}
\end{equation}


\begin{align*}
    &\text{(a) } x[n - 3] \\
    &\text{(b) } x[n + 4] \\
    &\text{(c) } x[-n] \\
    &\text{(d) } x[-n + 2] \\
    &\text{(e) } x[-n - 2]
\end{align*}


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = np.arange(-10,10)

def s(x):
    return np.where((x < -2) | (x > 4), 0, x)


def v(x):
    return np.where((x < 3), 0, x)
```

```python
fig, axs = plt.subplots(5, 1, figsize=(4, 15)) 

axs[0].stem(n, s(n - 3), basefmt=' ', linefmt='b-', label="$s[n - 3]$")
axs[0].set_xticks(n)
axs[0].set_xlabel('$n$')
axs[0].legend()

axs[1].stem(n, s(n + 4), basefmt=' ', linefmt='c-', label="$s[n + 4]$")
axs[1].set_xticks(n)
axs[1].set_xlabel('$n$')
axs[1].legend()

axs[2].stem(n, s(-n), basefmt=' ', linefmt='r-', label="$s[-n]$")
axs[2].set_xticks(n)
axs[2].set_xlabel('$n$')
axs[2].legend()

axs[3].stem(n, s(-n + 2), basefmt=' ', linefmt='g-', label="$s[-n + 2]$")
axs[3].set_xticks(n)
axs[3].set_xlabel('$n$')
axs[3].legend()

axs[4].stem(n, s(-n - 2), basefmt=' ', linefmt='m-', label="$s[-n - 2]$")
axs[4].set_xticks(n)
axs[4].set_xlabel('$n$')
axs[4].legend()

plt.tight_layout()
plt.show()
```

Reposta da 1:

\begin{align*}
s[n - 3]: & \quad n \leq 0 \text{ e } n \geq 8 \\
s[n + 4]: & \quad n \leq -6 \\
s[-n]: & \quad n \geq 5 \text{ e } n \leq -3 \\
s[-n + 2]: & \quad n \leq -2 \text{ e } n \geq 4 \\
s[-n - 2]: & \quad n \geq 3 \text{ e } n \leq -5
\end{align*}


```python
fig, axv = plt.subplots(5, 1, figsize=(4, 15)) 

axv[0].stem(n, v(1 - n), basefmt=' ', linefmt='b-', label="$v[1 - n]$")
axv[0].set_xticks(n)
axv[0].set_xlabel('$n$')
axv[0].legend()

axv[1].stem(n, v(1 - n) + v(2 - n), basefmt=' ', linefmt='c-', label="$v[1 - n] + v[2 - n]$")
axv[1].set_xticks(n)
axv[1].set_xlabel('$n$')
axv[1].legend()

axv[2].stem(n, v(1 - n) * v(2 - n), basefmt=' ', linefmt='r-', label="$v[1 - n] * v[2 - n]$")
axv[2].set_xticks(n)
axv[2].set_xlabel('$n$')
axv[2].legend()

axv[3].stem(n, v(3 * n), basefmt=' ', linefmt='g-', label="$v[3 * n]$")
axv[3].set_xticks(n)
axv[3].set_xlabel('$n$')
axv[3].legend()

axv[4].stem(n, v(n / 3), basefmt=' ', linefmt='m-', label="$v[n / 3]$")
axv[4].set_xticks(n)
axv[4].set_xlabel('$n$')
axv[4].legend()

plt.tight_layout()
plt.show()

```
