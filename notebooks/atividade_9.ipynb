Atividade para verificar as componentes pares e impares de um sinal discreto

```python
# Importar as bibliotecas necessárias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo as samples

# discreto, usado na q1
n = np.arange(-3, 3+1)

# continuo, usado na q2
t = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# discreto, usado na q3
x1 = np.arange(1, 10 + 1, 2)
x2 = np.arange(2, 12, 2)
```

```python
# # Função para calcular a componente par
# def pares(x):
#     N = len(x) 
#     x_par = np.zeros(N)
#     for n in range(N):
#         x_par[n] = (x[n] + x[N-n-1]) / 2

#     return x_par

# # Função para calcular a componente ímpar
# def impares(x):
#     N = len(x) 
#     x_impar = np.zeros(N)
#     for n in range(N):
#         x_impar[n] = (x[n] - x[N-n-1]) / 2

#     return x_impar

# Função para calcular a componente par (refatorada)
def pares(x):
    return (x + x[::-1]) / 2

# Função para calcular a componente ímpar (refatorada)
def impares(x):
    return (x - x[::-1]) / 2
```

### Questão 1: Calcule as componentes par e ímpar de um sinal discreto simples:
- (a) Defina um sinal discreto x[n] como uma lista de valores, por exemplo, x[n] = [1, 2, 3, 4, 5].

- (b) Implemente uma função para calcular a componente par xpar[n] do sinal.

- (c) Implemente uma função para calcular a componente ímpar xímpar[n] do sinal.

- (d) Plote os sinais originais, a componente par e a componente ímpar.

\begin{equation}
q1[n]=
\begin{cases} 
1 , \text{para } n ≥ 0 \\
0 , \text{para } n < 0 
\end{cases}
\end{equation}

```python
def q1(x):
    return np.where((x < 0), 0, 1)

f = q1(n)
```

```python
q1 = f
q1_par = pares(f)
q1_impar = impares(f)

fig, axs = plt.subplots(3, 1, figsize=(15, 15))

axs[0].stem(n, q1, basefmt=" ", linefmt='b-',)
axs[0].set_title('Sinal Original')
axs[0].set_xlabel('$n$')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)

axs[1].stem(n, q1_par, basefmt=" ", linefmt='g-',)
axs[1].set_title('Componentes Pares')
axs[1].set_xlabel('$n$')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)

axs[2].stem(n, q1_impar, basefmt=" ", linefmt='r-',)
axs[2].set_title('Componentes Ímpares')
axs[2].set_xlabel('$n$')
axs[2].set_ylabel('Amplitude')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

### Questão 2: Calcule as componentes par e ímpar de um sinal contínuo:

- (a) Defina um sinal contínuo x(t) = sin(t) + cos(2t) no intervalo de t = −2π a t = 2π.

- (b) Implemente funções para calcular as componentes par e ímpar do sinal contínuo.

- (c) Plote os sinais originais, a componente par e a componente ímpar.

\begin{equation}
q2(t) = \sin(t) + \cos(2t)
\end{equation}

```python
def q2(x): 
    return np.sin(x) + np.cos(2 * x)

g = q2(t)
```

```python
q2 = g
q2_par = pares(g)
q2_impar = impares(g)

fig, axs = plt.subplots(3, 1, figsize=(15, 15))

axs[0].plot(t, q2_par, 'g-', label='Componentes Pares')
axs[0].plot(t, q2_impar, 'r-', label='Componentes Ímpares')
axs[0].plot(t, q2, '--', label='Sinal Original')
axs[0].set_title('Sinal Original e suas componentes')
axs[0].set_xlabel('$t$')
axs[0].set_ylabel('Amplitude')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(t, q2_par, 'g-', label='Componentes Pares')
axs[1].set_title('Componentes Pares')
axs[1].set_xlabel('$t$')
axs[1].set_ylabel('Amplitude')
axs[1].grid(True)
axs[1].legend()


axs[2].plot(t, q2_impar, 'r-', label='Componentes Ímpares')
axs[2].set_title('Componentes Ímpares')
axs[2].set_xlabel('$t$')
axs[2].set_ylabel('Amplitude')
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```
